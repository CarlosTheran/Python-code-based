```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

```python
# loading images
img1 = cv2.imread('/Users/carlos-theran/Documents/famu-spring22/Mat4906-ML/images_for_python/forest.jpg')
img2 = cv2.imread('/Users/carlos-theran/Documents/famu-spring22/Mat4906-ML/images_for_python/people-walking.jpg')
type(img2)
print('showing the size of image 1', img1.shape)
print('showing the size of image 2', img2.shape)
img1_reshaped = cv2.resize(img1,(300,300),interpolation=cv2.INTER_AREA)
img2_reshaped = cv2.resize(img2,(300,300),interpolation=cv2.INTER_AREA)
print('showing size of img2_reshaped', img2_reshaped.shape)
```

```python
#printing the images
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img1_reshaped)
plt.subplot(122)
plt.imshow(img2_reshaped)
```

```python
# printing RGB channels

plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img1_reshaped[:,:,0])
plt.gca().set_title('Red Channel')
plt.subplot(132)
plt.imshow(img1_reshaped[:,:,1])
plt.gca().set_title('Geen Channel')
plt.subplot(133)
plt.imshow(img1_reshaped[:,:,2])
plt.gca().set_title('Blue Channel')
```

```python
dst = cv2.addWeighted(img2_reshaped,0.5,img1_reshaped,0.3,0.1)
plt.figure(figsize=(10,10))
plt.imshow(dst)
```

```python
dst_subt = cv2.subtract(img1_reshaped,img2_reshaped)
plt.figure(figsize=(10,10))
plt.imshow(dst_subt)
```

```python
img_cat = cv2.imread('/Users/carlos-theran/Documents/famu-spring22/Mat4906-ML/images_for_python/cat.jpg')
negative = 255 - img_cat
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img_cat)
plt.gca().set_title("Original Image")
plt.axis("off")
plt.subplot(122)
plt.imshow(negative)
plt.gca().set_title("Negative Image")
plt.axis("off")
```

```python
img_blur = cv2.blur(img_cat,(80,80))
img_filter = cv2.boxFilter(img_cat,-5,(30,30),normalize=True)
img_gaussianfilter = cv2.GaussianBlur(img_cat,(41,41),cv2.BORDER_DEFAULT)

plt.figure(figsize=(20,20))
plt.subplot(141)
plt.imshow(img_cat)
plt.gca().set_title("Original Image")
plt.subplot(142)
plt.imshow(img_blur)
plt.gca().set_title("Blur Image")
plt.subplot(143)
plt.imshow(img_filter)
plt.gca().set_title("Filter Image")
plt.subplot(144)
plt.imshow(img_gaussianfilter)
plt.gca().set_title("Gaussian Image")
```

```python
img_median = cv2.medianBlur(img_cat,31)
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img_cat)
plt.gca().set_title("Original Image")
plt.axis("off")
plt.subplot(122)
plt.imshow(img_median)
plt.gca().set_title("Blur image using MedianBlur")
plt.axis("off")
```

```python
img_cat_hsv = cv2.cvtColor(img_cat,cv2.cv2.COLOR_RGB2HSV)
plt.figure(figsize=(10,10))
plt.imshow(img_cat_hsv)
```

```python
print(img_cat[800,1000,:])
upper_limit = (119,119,13)#(13,119,119)
lower_limit = (191,191,61)#(61,119,119)
mask = cv2.inRange(img_cat,upper_limit,lower_limit)
result = cv2.bitwise_and(img_cat,img_cat,mask=mask)
plt.figure(figsize=(20,20))
plt.subplot(121)
plt.imshow(mask, cmap="gray")
plt.subplot(122)
plt.imshow(result)
```
