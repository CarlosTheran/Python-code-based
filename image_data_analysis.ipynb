```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

```python
# loading images
img1 = cv2.imread('/Users/carlos-theran/Documents/famu-spring22/Mat4906-ML/images_for_python/forest.jpg')
img2 = cv2.imread('/Users/carlos-theran/Documents/famu-spring22/Mat4906-ML/images_for_python/people-walking.jpg')
type(img2)
print('showing the size of image 1', img1.shape)
print('showing the size of image 2', img2.shape)
img1_reshaped = cv2.resize(img1,(300,300),interpolation=cv2.INTER_AREA)
img2_reshaped = cv2.resize(img2,(300,300),interpolation=cv2.INTER_AREA)
print('showing size of img2_reshaped', img2_reshaped.shape)
```

```python
#printing the images
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img1_reshaped)
plt.subplot(122)
plt.imshow(img2_reshaped)
```

```python
# printing RGB channels

plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img1_reshaped[:,:,0])
plt.gca().set_title('Red Channel')
plt.subplot(132)
plt.imshow(img1_reshaped[:,:,1])
plt.gca().set_title('Geen Channel')
plt.subplot(133)
plt.imshow(img1_reshaped[:,:,2])
plt.gca().set_title('Blue Channel')
```
