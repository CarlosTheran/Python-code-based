```python
import pandas
import numpy  as np
import seaborn as sns
```

```python
data = pandas.read_csv("/Users/carlos-theran/Downloads/nba.csv")
data.head()
```

```python
#Creating Some Basic plots using seaborn

# Selecting style as white,
# dark, whitegrid, darkgrid 
# or ticks
sns.set( style = "white" )
   
# Generate a random univariate 
# dataset
rs = np.random.RandomState( 10 )
d = rs.normal( size = 50 )
   
# Plot a simple histogram and kde 
# with binsize determined automatically
sns.distplot(d, kde = True, color = "g")
```
